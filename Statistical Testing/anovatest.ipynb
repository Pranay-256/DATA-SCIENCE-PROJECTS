```python
import seaborn as sns
import pandas as pd
from scipy.stats import f_oneway
```

```python
df = sns.load_dataset('titanic')
```

```python
df
```

```python
df = df[['age','pclass']].dropna() #remove null values from age and pclass
```

```python
df['pclass'].unique() #gives all the classes in pclass
```

```python
class_1 = df[df['pclass']==1]['age'] # Selects only the 'age' values for passengers whose 'pclass' is equal to 1 (First Class) and stores them in the variable class_1.
class_2 = df[df['pclass']==2]['age']
class_3 = df[df['pclass']==3]['age']
class_1
```

```python
f_stats , p_value = f_oneway(class_1,class_2,class_3)
```

```python
f_stats
```

```python
p_value
```

```python
alpha = 0.05
```

```python
if p_value<alpha:
    print("reject the null hypothesis and there is a significant difference between at least one passenger class")
else:
    print("there is no significant difference")
```
