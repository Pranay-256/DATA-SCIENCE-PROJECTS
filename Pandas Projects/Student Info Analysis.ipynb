### Create a new column score_level
"High" if total_score â‰¥ 90 , "Medium" if 70 â‰¤ total_score < 90, "Low" if total_score < 70

```python
def level(x):
  if x >= 90:
    return "High"
  elif x>= 70 and x < 90:
    return "Medium"
  else:
    return "Low"
```

```python
df["Score_level"] = df["total_score"].apply(level)
df
```
